# Filtering and performance metric calculations
The following filters are applied
- IEC
- STC
- (nighttime)
- outlier
- clear-sky
- irradiance thresholds


The input data (e.g., "helsinki_prefiltered_filled_missing_data_cell_T_rc_UTC00.csv") were created in "data_pre_processing.ipynb" notebook

```python
# Filters and corresponding names
poa_lims = [0, 5, 20, 50, 100, 200, 500, 800]
power_vs_poas = [False, True]
clearskies = [False, True]
filters = [{'power_vs_poa':p_vs_poa, 'clearsky':clearsky, 'poa_threshold':(poa_l, 2000)}
            for poa_l in poa_lims
            for p_vs_poa in power_vs_poas
            for clearsky in clearskies
           ]
def get_filter_name(p_vs_poa, clearsky, poa_l):
    pv_vs_poa_name = 'PvsPOA_' if p_vs_poa else ''
    clearsky_name = 'CS20_' if clearsky else ''
    poa_l_name = 'POA'+str(poa_l)
    return pv_vs_poa_name + clearsky_name + poa_l_name

filter_names = [get_filter_name(f['power_vs_poa'], f['clearsky'], f['poa_threshold'][0]) for f in filters]

# Performance metrics with different aggregation intervals
metrics = {'power': np.array(['insta', 'D', 'W', 'ME']), 'pi': np.array(['insta', 'D', 'W', 'ME']), 'pr': np.array(['D', 'W', 'ME']),
          'prt': np.array(['D', 'W', 'ME']), 'pvusa': np.array(['D', 'W', 'ME']), 'huld': np.array(['D', 'W', 'ME'])}
```

```python
DATA_PATH = 'Datasets'
```

```python
path = os.path.join(DATA_PATH, 'kuopio_performance_metrics')  # Directory of performance datasets
paths = glob.glob(os.path.join(path, "*"))
paths = [p for p in paths if os.path.basename(p) != "insta"]

filename = 'PLRs_kuopio.csv'
post_filter = None
folder = os.path.join(DATA_PATH, 'PLRs_yoy_uncertainty')
for path in paths:  # Iterate through every folder (insta, D, W, M)
    all_files = glob.glob(os.path.join(path, "*.csv"))
    for file in all_files:
        POA_u=None
        POA_l=None
        set_POA_thresholds = False
        if 'IEC' in file:
            POA_u=1500
            POA_l=-6
            set_POA_thresholds = True
        elif 'STC' in file:
            POA_u=1050
            POA_l=950
            set_POA_thresholds = True
            
        linear_regression(filename, folder, filepath=file, post_filter=post_filter, POA_u=POA_u, POA_l=POA_l, set_POA_thresholds=set_POA_thresholds)
        robust_regression(filename, folder, filepath=file, post_filter=post_filter, POA_u=POA_u, POA_l=POA_l, set_POA_thresholds=set_POA_thresholds)
        yoy(filename, folder, filepath=file, post_filter=post_filter, POA_u=POA_u, POA_l=POA_l, set_POA_thresholds=set_POA_thresholds)
```
